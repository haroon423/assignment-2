```python
import pandas as pd
data = pd.read_csv("pokemon.csv")
data.head()
data.tail(n=10)
```

```python

data["HP"].mean()
```

```python
HP_mean =data["HP"].mean()
def set_HP(val):
    if val < HP_mean:
        return "HP Low"
    else:
        return "HP High"
    
    
    
    
    
    
    
   
```

```python
data["HP_High_Low"] = data["HP"].apply(set_HP)
```

```python
data
```

```python
data["Speed"].mean()
```

```python
Speed_mean=data["Speed"].mean()
def set_Speed(val):
    if val < Speed_mean:
        return "Speed Low"
    else:
        return "Speed High"
    
  
    
```

```python
data["Speed_High_=Low"] =data["Speed"].apply(set_Speed)
```
